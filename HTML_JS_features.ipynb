# HTML and JavaScript based Features
Many features can be extracted that come under this category. Out of them, below mentioned were considered for this project.

1. IFrame Redirection
2. Status Bar Customization
3. Disabling Right Click
4. Website Forwarding

```python
import requests
from bs4 import BeautifulSoup
import pandas as pd
```

```python
# 1.IFrame Redirection (iFrame)
def iframe(url):
    try:
        response = requests.get(url, timeout=120)
    except:
        return 2
    try:
        soup = BeautifulSoup(response.text, 'html.parser')
    except MemoryError:
        return 2
    iframes = soup.find_all('iframe')
    if not iframes:
        return 0
    for iframe in iframes:
        if 'frameborder' not in iframe.attrs or iframe.get('frameborder') == '0':
            return 1
        if 'style' in iframe.attrs and 'display:none' in iframe.get('style'):
            return 1
    else:
        return 0
# 0->No suspicious iframes
# 1->Hidden or borderless iframe found
# 2->Request blocked or timed out
```

```python
# 2.Checks the effect of mouse over on status bar (Mouse_Over)
def mouseOver(url): 
    try:
        response = requests.get(url, timeout=120)
    except:
        return 2
    try:
        soup = BeautifulSoup(response.text, 'html.parser')
    except MemoryError:
        return 2
    elements = soup.find_all(onmouseover=True)
    if not elements:
        return 0
    for element in elements:
        onmouseover = element.get('onmouseover')
        if 'window.status=' in onmouseover:
            return 1
    else:
        return 0
# 0->No status bar modification
# 1->Status bar modified
# 2->Request blocked or timed out
```

```python
# 3.Checks the status of the right click attribute (Right_Click)
def rightClick(url):
    try:
        response = requests.get(url, timeout=120)
    except:
        return 2
    try:
        soup = BeautifulSoup(response.text, 'html.parser')
    except MemoryError:
        return 2
    scripts = soup.find_all('script')
    for script in scripts:
        if 'event.preventDefault()' in script.text or 'return false' in script.text:
            return 1
    else:
        return 0
# 0->Right click allowed
# 1->Right click potentially blocked
# 2->Request blocked or timed out
```

```python
# 4.Checks the number of forwardings (Web_Forwards)    
def forwarding(url):
    try:
        response = requests.get(url, timeout=120)
    except:
        return -1
    return len(response.history)
# -1->Request blocked or timed out
# Whole numbers->Number of forwards
```

```python
#Function to extract features
def featureExtraction(url,label):
    features = []
    # HTML & Javascript based features (4)
    ifr = iframe(url)
    features.append(ifr)
    print("Check 14 of 17 complete.", end="\r")
    if ifr == 2:
        features.append(2)
        print("Check 15 of 17 complete.", end="\r")
        features.append(2)
        print("Check 16 of 17 complete.", end="\r")
        features.append(-1)
        print("Check 17 of 17 complete.", end="\r")
    else: 
        features.append(mouseOver(url))
        print("Check 15 of 17 complete.", end="\r")
        features.append(rightClick(url))
        print("Check 16 of 17 complete.", end="\r")
        features.append(forwarding(url))
        print("Check 17 of 17 complete.", end="\r")
    return features
```

```python
legi_features = []
label = 0
df_phish = pd.read_csv(r'data\selected_phishing_urls.csv')
for i in range(100, 5000):
    url = df_phish['URLs'][i]
    print("Extracting features for url number "+str(i)+" : "+url)
    legi_features.append(featureExtraction(url,label))
```

```python
feature_names = ['iFrame', 'Mouse_Over','Right_Click', 'Web_Forwards']

legitimate = pd.DataFrame(legi_features, columns= feature_names)
legitimate.head(10)
```
