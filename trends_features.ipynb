```python
import re
import requests
from pytrends.exceptions import ResponseError
from pytrends.request import TrendReq
import time
import whois
import urllib
import urllib.request
from datetime import datetime
import pandas as pd
from urllib.parse import urlparse,urlencode
```

```python
df_phish1 = pd.read_csv(r'feature_extracted_data\phishing_urls_Trends.csv')
df_phish1.drop_duplicates(subset="Domain", keep="first", inplace=True)
known_domains = df_phish1['Domain'].to_list()
known_interests = df_phish1['Comb_5y_interest'].to_list()
```

```python
# 2.Web traffic (Web_Traffic)
pytrend = TrendReq()
# Get the data for the keyword
def automate_trends(url):
    keyword = urlparse(url).netloc
    if re.match(r"^www.",keyword):
        keyword = keyword.replace("www.","")
    try:
        interest_pos = known_domains.index(keyword)
        req_interest = known_interests[interest_pos]
        print("Interest known already")
        return req_interest
    except:
        try:
            pytrend.build_payload(kw_list=[keyword])
            data = pytrend.interest_over_time()
        except:
            time.sleep(120)
            try:
                response = pytrend.build_payload(kw_list=[keyword])
                data = pytrend.interest_over_time()
            except ResponseError:
                return 0
        try:
            interest = data[keyword].tolist()
        except:
            interest = [0]
        total_interest = sum(interest)
        return total_interest
```

```python
def featureExtraction(url,label):
    features = []
    domain = urlparse(url).netloc
    if re.match(r"^www.",domain):
        domain = domain.replace("www.","")
    features.append(domain)
    features.append(automate_trends(url))
    print("Check 11 of 17 complete.", end="\r")
    return features
```

```python
legi_features = []
label = 0
df_phish = pd.read_csv(r'data\selected_phishing_urls.csv')
for i in range(2741, 5000):
    url = df_phish['URLs'][i]
    print("Extracting features for url number "+str(i)+" : "+url)
    legi_features.append(featureExtraction(url,label))
```

```python
feature_names = ['Domain','Comb_5y_interest']
legitimate = pd.DataFrame(legi_features, columns= feature_names)
legitimate.head(10)
```
