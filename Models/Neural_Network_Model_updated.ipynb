```python
import pandas as pd
import numpy as np
import tensorflow as tf
from keras.preprocessing import text
from keras.layers import Dense
from keras.layers import Embedding
from keras.models import Sequential
from sklearn.model_selection import train_test_split
```

```python
df = pd.read_csv('C:\\Users\\aaron\\Documents\\Project_Work\\feature_extracted_data\\train.csv')
```

```python
df
```

```python
tokens = text.Tokenizer()
tokens1 = text.Tokenizer()
tokens.fit_on_texts(df['URL'])
tokens1.fit_on_texts(df['Domain'])

df['URL'] = tokens.texts_to_matrix(df['URL'])
df['Domain'] = tokens.texts_to_matrix(df['Domain'])
```

```python
df
```

```python
x = df.iloc[:,0:18]
y = df.iloc[:,18]
```

```python
x.shape
```

```python
y.shape
```

```python
seq = Sequential()
seq.add(Dense(8, activation='relu', input_shape=(None,1,18)))
seq.add(Dense(8, activation='relu'))                  # Relu is used as it returns only positive values
seq.add(Dense(1, activation='sigmoid'))   #sigmoid is used as we want the output to be within 0 and 1
seq.compile(loss='binary_crossentropy',               # binary_crossentropy is used since the output is a binary value (0 or 1 || M or B)
optimizer='RMSPROP',                                        
metrics=['accuracy'])
```

```python
seq.fit(x, y ,epochs=200, batch_size=16, verbose=1) 
```

```python
y_pred = seq.predict(x)
```

# CONFUSION MATRIX

```python
import tensorflow as tf
from sklearn.metrics import confusion_matrix

confusion_matrix = tf.math.confusion_matrix(labels=y,predictions=y_pred).numpy()
```

```python
confusion_matrix
```
